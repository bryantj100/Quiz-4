```python
# Generates random sequences of length N and find the longest stretch C residues, prints the sequence, excursion plots, a histogram of the max scores.
# Also gives the maximal segmental score, and the index for the score

import random
import matplotlib.pyplot as plt


max_score_list = []

for i in range (10): #finds the maximal segmental score and index for that score for R sequences specified in the range
    seq = ""
    while len(seq) < 100: # specifies the length of the sequence:
        r = random.random() # Random sequence generator. Selects r such that 0<=r<=
        if r<=0.26:
            seq = seq + 'C'
        elif r<=0.55:
            seq = seq + 'P'
        elif r<=1.0:
            seq = seq + 'H'


x_axis = [0]
score_list = [0]
score = 0
new_score = 0


for i in range(0,len(seq)): # sets the score to +1 for C, -2 for P, and -3 for H
    if seq[i] == 'C':
        score = 1
    elif seq[i] == 'P':
        score = -2
    else:
        score =-3

    new_score = max(new_score + score, 0)
    score_list.append(new_score)


k = 0
while k < len(seq): #creates the x axis
    k = k+1
    x_axis.append(k)

max_score_list.append(max(score_list))


print("The maximal segmental score for my sequence is: ",  max(score_list))
print("The index for the maximal segmental score is: ",  score_list.index(max(score_list)))
plt.plot(x_axis,score_list)

print(max_score_list)
print("Max value is: ", max(max_score_list))
plt.hist(max_score_list)
plt.title("Distribution of max scores")
plt.show()
```
